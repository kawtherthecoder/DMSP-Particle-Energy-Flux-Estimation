```python
import numpy as np
import csv
import matplotlib.pyplot as plt
```

```python
arr = np.loadtxt("./CSCI5922 Project - Loss Epochs.csv",
                 delimiter=",", dtype=float)

print(arr.shape)
```

```python
inds = np.array([x for x in range(0,12) if x % 2 == 0])

print(inds)

arr[arr == -999] = np.nan

plt.figure(figsize=(6,6))
for i in inds:
    plt.semilogy(arr[i,:],linewidth=2)

plt.xticks((range(0,20,4)));
plt.xlabel('Epoch')
plt.ylabel('MSE')
plt.legend(['AR16 ACRELU','AR8-32 ACRELU','AR16-24 ACRELU','AR16 ACLRELU','AR8-32 ACLRELU','AC16-24 ACLRELU']);
plt.title('Optimizer = Adam');
```
